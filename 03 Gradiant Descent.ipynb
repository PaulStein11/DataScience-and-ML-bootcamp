# Example 1 - Simple cost function
## $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1
```

```python
#Make data
x_1 = np.linspace(start=-3, stop= 3, num=500)
```

```python
#Plot
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.xlim(-3, 3)
plt.ylim(0, 8)
plt.plot(x_1, f(x_1))
plt.show()
```

## SLOPE AND DERIVATIVES

Challenge: Create a python function with the derivative of $f(x)$ called $df(x)$

```python
def df(x):
    return 2*x + 1
```

```python
#Plot function and derivative side by side

plt.figure(figsize= [15, 5])

# 1.Chart: Cost function
plt.subplot(1, 2, 1)

plt.title("Cost function", fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.plot(x_1, f(x_1), color="blue", linewidth=3)

# 2.Chart: Derivative
plt.subplot(1, 2, 2)
plt.grid()
plt.plot(x_1, df(x_1), color="skyblue", linewidth=5)

plt.title("Slope of the cost function", fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.show()
```

## Python Loops and Gradient Descend

```python
# Python For Loop
for n in range(5):
    print("Hello world", n)
print("End of for loop")
```

```python
# Python while Loop

counter = 0
while counter <= 7:
    print("Counter...", counter)
    counter = counter + 1;
print("Finished or not, here I come!")
```

```python
# Gradient descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

## creation of lists to append values
x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    step_size = abs(new_x - previous_x)
    # print(step_size)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        print('Loop ran this many times:' , n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

```python
#Superimpose the gradient descent calculation

plt.figure(figsize= [20, 5])

# 1.Chart: Cost function
plt.subplot(1, 3, 1)

plt.title("Cost function", fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.plot(x_1, f(x_1), color="blue", linewidth=3, alpha=0.6)

#Conver values to make it suitable with plt taht requires an int
values = np.array(x_list)
plt.scatter(x_list, f(values), color="red", s=100, alpha=0.6)

# 2.Chart: Derivative
plt.subplot(1, 3, 2)
plt.grid()

plt.title("Slope of the cost function", fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color="skyblue", linewidth=5, alpha=0.5)
plt.scatter(x_list, slope_list, color="red", s=100, alpha=0.5)


# 3.Chart: Derivative (Close Up)
plt.subplot(1, 3, 3)
plt.grid()

plt.title("Gradient descent (close up)", fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)

plt.plot(x_1, df(x_1), color="skyblue", linewidth=5, alpha=0.5)
plt.scatter(x_list, slope_list, color="red", s=100, alpha=0.5)



plt.show()
```
